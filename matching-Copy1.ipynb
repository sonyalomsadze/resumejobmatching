```python
import pandas as pd
import numpy as np

job_df = pd.read_csv('./data/dice_com-job_us_sample.csv')
user_df = pd.read_csv('./data/survey_results_public.csv')
```

```python
job_df.head(2)
```

```python
user_df.head(2)
```

```python
job_df['meta_data'] = job_df['jobtitle'] + ' ' + job_df['skills']
job_df['type'] = 'job' + '_' +  job_df['uniq_id']
job_df = job_df[['type', 'meta_data']]
print(job_df.info())
job_df.T
```

```python
user_df = user_df[~(user_df.DevType.isna())]
user_df['meta_data'] = user_df['DevType'] + user_df['LanguageWorkedWith']
user_df['type'] = user_df['Respondent']
user_df = user_df[['type', 'meta_data']]
user_df = user_df[~user_df.meta_data.isna()]
print(user_df.info())

user_df.T
```

```python
df = pd.concat([job_df, user_df])
df =  df[~df.meta_data.isna()]
df
```

```python
df[:21960]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

X = df[:30000].meta_data

vectorizer =  CountVectorizer().fit_transform(X)
vectors = vectorizer.toarray()
csim = cosine_similarity(vectors)
```

```python
match = pd.DataFrame(csim)
match = match.iloc[:21957, 21957:]
match
```

```python
nlargest = 5
order = np.argsort(-match.values, axis=1)[:, :nlargest]
result = pd.DataFrame(match.columns[order], 
                      columns=['top{}'.format(i) for i in range(1, nlargest+1)],
                      index=match.index)

print(result)
```
